```python
import matplotlib.pyplot as plt
```

```python
from skimage.filters import roberts
import skimage.filters
```

```python
roberts_img = roberts(cv2.imread("download.jpeg",0))
```

```python
roberts_img
```

```python
plt.figure(figsize=(20,25))
plt.imshow(roberts_img, cmap='Blues')
plt.show()
```

```python
plt.imshow(roberts_img, cmap='gray')
plt.show()
```

```python
threshold = skimage.filters.apply_hysteresis_threshold(roberts_img, np.mean(roberts_img), np.mean(roberts_img))
```

```python
threshold
```

```python
plt.imshow(threshold)
plt.show()
```

```python
plt.imshow(roberts_img, cmap='hsv_r')
plt.show()
```
